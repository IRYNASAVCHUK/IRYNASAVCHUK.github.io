```python
# Set general plot style
sns.set_theme(style="whitegrid")

# Define numerical and categorical columns
numerical_cols = ["Victim Age", "Treatment Cost"]
categorical_cols = [
    "Bite Type", "Victim Relationship", "Bite Location",
    "Bite Severity", "Bite Circumstance", "Controlled By"
]

# Plot histograms for numerical features
fig, axes = plt.subplots(1, len(numerical_cols), figsize=(12, 5))
for i, col in enumerate(numerical_cols):
    sns.histplot(df[col].dropna(), bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()

# Plot bar charts for categorical features
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    value_counts = df[col].dropna().value_counts()
    sns.barplot(x=value_counts.index, y=value_counts.values, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xticks(range(len(value_counts.index)))  # Set tick positions
    axes[i].set_xticklabels(value_counts.index, rotation=90)  # Set labels with rotation

plt.tight_layout()
plt.show()
```

```python
# Rimuoviamo spazi indesiderati nei nomi delle colonne
df.columns = df.columns.str.strip()

# Convertiamo le date, facendo attenzione ai formati
df["Incident Date"] = pd.to_datetime(df["Incident Date"], errors='coerce', format="%m/%d/%Y %I:%M:%S %p")
df["Date Reported"] = pd.to_datetime(df["Date Reported"], errors='coerce', format="%m/%d/%Y %I:%M:%S %p")

# Pulizia della colonna "Victim Age" (convertire in numerico e sostituire NaN con la mediana)
df["Victim Age"] = pd.to_numeric(df["Victim Age"], errors='coerce')
df["Victim Age"].fillna(df["Victim Age"].median(), inplace=True)

# Pulizia della colonna "Treatment Cost" (convertire in numerico e sostituire NaN con 0)
df["Treatment Cost"] = pd.to_numeric(df["Treatment Cost"], errors='coerce')
df["Treatment Cost"].fillna(0, inplace=True)

# Sostituzione valori mancanti nelle colonne categoriche con "Unknown"
categorical_columns = ["Victim Relationship", "Bite Location", "Bite Severity", 
                       "Bite Circumstance", "Controlled By", "Incident Location"]
df[categorical_columns] = df[categorical_columns].fillna("Unknown")

# Rimozione duplicati
df.drop_duplicates(inplace=True)

# Controllo finale
print("Dati dopo la pulizia:")
print(df.info())
print(df.head())
```

```python
# Contiamo i casi con lo stesso "Bite Number" ma dettagli diversi
duplicates = df[df.duplicated(subset=["Bite Number"], keep=False)]

print("Possibili casi di più vittime per lo stesso morso:")
print(duplicates.sort_values(by="Bite Number"))

# Rimuoviamo solo i duplicati perfetti (tutte le colonne uguali)
df = df.drop_duplicates()

# Verifichiamo quanti record restano dopo la pulizia
print(f"Numero di righe dopo la rimozione di duplicati esatti: {len(df)}")
```

```python
# Salviamo il dataset pulito in un nuovo file CSV
output_path = "data/Dog_Bite_Cleaned.csv"
df.to_csv(output_path, index=False)

print(f"File pulito salvato in: {output_path}")
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError
df = df.head(100)
# Initialize geolocator
geolocator = Nominatim(user_agent="your_app_name")

def get_coordinates(address):
    try:
        location = geolocator.geocode(address, timeout=10)
        if location:
            return location.latitude, location.longitude
    except (GeocoderTimedOut, GeocoderServiceError) as e:
        print(f"Error geocoding {address}: {e}")
    return None, None


# Adding index to see progress
df["Latitude"], df["Longitude"] = zip(*df.apply(lambda row: get_coordinates(row["Incident Location"]), axis=1))

```

```python
import folium

# Center map on Dallas
map_dallas = folium.Map(location=[32.7767, -96.7970], zoom_start=10)

# Add incident locations as markers
for index, row in df.iterrows():
    if pd.notna(row["Latitude"]) and pd.notna(row["Longitude"]):
        folium.Marker(
            location=[row["Latitude"], row["Longitude"]],
            popup=row["Incident Location"],
            icon=folium.Icon(color="red", icon="info-sign"),
        ).add_to(map_dallas)

# Save or display the map
map_dallas.save("dallas_incidents_map.html")
```
